```python
from sentence_transformers import SentenceTransformer
import chromadb
chroma_client = chromadb.PersistentClient(path="./chromadb")
collection = chroma_client.get_or_create_collection(name="tweet_embeddings")
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-mpnet-base-v2')
```

```python
import json
with open('all_tweets.json', 'r', encoding='utf-8') as file:
        all_tweet = json.load(file)
print(len(all_tweet))
```

```python
for index, tweet in enumerate(all_tweet):
    print(index + 1, "/", len(all_tweet))
    tweet['embedding'] = model.encode(tweet['full_text'])
```

```python
for index, tweet in enumerate(all_tweet):
    print(index + 1, "/", len(all_tweet))
    meta = {
        'created_at': tweet['created_at'],
        'id': tweet['id'],
        'full_text': tweet['full_text']
    }
    res = collection.upsert(
        documents=[tweet['full_text']],
        embeddings=[tweet['embedding']],
        metadatas=[meta],
        ids=[tweet['id']]
    )
    print(res)
```
